# From the accuracy we can say that the model is overfitted to avoid the overfitting we use some regularozation method



```python
# Here we have l1 l2 regularization  

#it turn out they have diffrent but equally useful properties. for a partical standpoint,

#l1 tends to shirnk coefficient to zero ( 0 ). 

#l2 tends to shrink coefficents evenly.

#l1 is useful for feature selection, we can drop any variable associated with coefficients that go to zero.

```

```python
from sklearn import linear_model

lasso=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso.fit(x_train,y_train)
```

```python
lasso.score(x_test,y_test)
```

```python
lasso.score(x_train,y_train)
```
