```python
# Probability values  
lr.predict_proba (x_train) 
```

# After traing the model now we can test the model

```python
y_pred=lr.predict(x_test)
y_pred
```

# Let's the performance of our model - Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred),annot= True, cbar= False)
```

```python
#Encode the target variable

le = LabelEncoder()

y_test_enc = le.fit_transform(y_test) 
y_test_enc 
```

```python
proba = lr.predict_proba(x_test)[:,1]
proba 
```

```python
auc = roc_auc_score(y_test_enc, proba)
auc 
```

```python
roc_curve(y_true=y_test_enc, y_score=proba) 
fpr, tpr, threshold = roc_curve(y_test_enc, proba)  
fpr, tpr, threshold
```

```python
pd.DataFrame({'False Positive Rate': fpr,
             'True Positive Rate': tpr,
             'Threshold': threshold}) 
```

```python
#ROC Curve

plt.rcParams['figure.figsize'] = (5,3)
plt.rcParams['figure.dpi'] = 150
plt.rcParams['font.size'] = 6
plt.plot(fpr, tpr, label = '(AUC = %0.2f)' %auc)
plt.xlabel('False Positive rate (fpr)')
plt.ylabel('True Positive rate (tpr)')
plt.title('ROC Curve')
plt.legend() 
```
